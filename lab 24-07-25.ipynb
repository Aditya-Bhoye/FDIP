```python
import numpy as np
import matplotlib.pyplot as plt 
from skimage import img_as_uint
from skimage.io import imshow, imread
from skimage.color import rgb2hsv
from skimage.color import rgb2gray
```

```python
array_1=np.array([[255,0],[0,255]])
imshow(array_1,cmap='gray');
```

```python
array_2=np.array([[255,0,255],[0,255,0],[255,0,0]])
imshow(array_2,cmap='gray')
```

```python
array_spectrum = np.array([np.arange(0,255,17), 
                   np.arange(255,0,-17),
                   np.arange(0,255,17),
                   np.arange(255,0,-17)])
fig, ax = plt.subplots(1, 2, figsize=(12,4))
ax[0].imshow(array_spectrum, cmap = 'gray')
ax[0].set_title('Arange Generation')
ax[1].imshow(array_spectrum.T, cmap = 'gray')
ax[1].set_title('Transpose Generation');
```

```python
array_colors = np.array([[[255, 0, 0], 
                         [0, 255, 0],
                         [0, 0, 255]]])
imshow(array_colors)
```

```python
array_colors=np.array([[[0,255,0],
                       [255,0,0],[
                           0,0,255]]])
imshow(array_colors)
```

```python
bmw=imread('bmw.jpg')
imshow(bmw)
```

```python
bmw.shape
```

```python
fig, ax=plt.subplots(1,3,figsize=(6,4),sharey=True) 
ax[0].imshow(bmw[:, 0:250]) 
ax[0].set_title('First Split') 
ax[1].imshow(bmw[:, 250:500]) 
ax[1].set_title('Second Split') 
ax[2].imshow(bmw[:, 500:735]) 
ax[2].set_title('Third Split')
```

```python
imshow(bmw[200:600,200:600])
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12,4), sharey =True)
ax[0].imshow(bmw[:,:,0],cmap='Reds')
ax[0].set_title('Red')
ax[1].imshow(bmw[:,:,1],cmap='Greens')
ax[1].set_title('Green')
ax[2].imshow(bmw[:,:,2],cmap='Blues')
ax[2].set_title('Blue')
```

```python
bmw_hsv = rgb2hsv(bmw)
fig, ax =plt.subplots(1, 3, figsize=(12,4), sharey = True)
ax[0].imshow(bmw_hsv[:,:,0], cmap='hsv')
ax[0].set_title('Hue')
ax[1].imshow(bmw_hsv[:,:,1],cmap='gray')
ax[1].set_title('Saturation')
ax[2].imshow(bmw_hsv[:,:,2], cmap='gray')
ax[2].set_title('Value');
```

```python
bmw_gray = rgb2gray(bmw)
fig, ax = plt.subplots(1, 5, figsize=(17,6), sharey = True)
ax[0].imshow(bmw_gray, cmap = 'gray')
ax[0].set_title('Grayscale Original')
ax[1].imshow(img_as_uint(bmw_gray > 0.25), cmap = 'gray')
ax[1].set_title('Greater than 0.25')
ax[2].imshow(img_as_uint(bmw_gray > 0.50), cmap = 'gray')
ax[2].set_title('Greater than 0.50');
ax[3].imshow(img_as_uint(bmw_gray > 0.75), cmap = 'gray')
ax[3].set_title('Greater than 0.75');
ax[4].imshow(img_as_uint(bmw_gray > np.mean(bmw_gray)), cmap = 'gray')
ax[4].set_title('Greater than Mean');
```

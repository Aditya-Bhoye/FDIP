```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("capri.jpg", cv2.IMREAD_GRAYSCALE)
if img is None:
    raise FileNotFoundError("Image not found!")

plt.imshow(img,cmap='gray')
"""plt.title("Original Image")"""
plt.axis("off")
plt.show()
```

```python

C = 50
bright_up = np.clip(img.astype(np.int16) + C, 0, 255).astype(np.uint8)

plt.imshow(bright_up, cmap='gray')
plt.title("Brightness Increased")
plt.axis("off")
plt.show()
```

```python
bright_down = np.clip(img.astype(np.int16) - C, 0, 255).astype(np.uint8)

plt.imshow(bright_down, cmap='gray')
plt.title("Brightness Decreased")
plt.axis("off")
plt.show()
```

```python
neg = (255 - img).astype(np.uint8)

plt.imshow(neg, cmap='gray')
plt.title("Negative Image")
plt.axis("off")
plt.show()
```

```python
T = 128
thresh = np.where(img >= T, 255, 0).astype(np.uint8)

plt.imshow(thresh, cmap='gray')
plt.title("Thresholded Image")
plt.axis("off")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('bmv.jpg', cv2.IMREAD_GRAYSCALE)
if image is None:
    print("Image not found!")
else:
    plt.imshow(image, cmap='gray')
    plt.title("Original Image")
    plt.axis('off')
    plt.show()
```

```python
mean = 0
std_dev = 25
gaussian_noise = np.random.normal(mean, std_dev, image.shape)
gaussian_image = np.clip(image + gaussian_noise, 0, 255).astype(np.uint8)
plt.imshow(gaussian_image, cmap='gray')
plt.title("Gaussian Noisy Image")
plt.axis('off')
plt.show()
```

```python
plt.hist(gaussian_image.ravel(), bins=256, range=(0,255), color='blue')
plt.title("Histogram - Gaussian Noise")
plt.show()
```

```python
prob = 0.05
sp_image = np.copy(image)
num_salt = np.ceil(prob * image.size * 0.5)
coords = [np.random.randint(0, i-1, int(num_salt)) for i in image.shape]
sp_image[coords[0], coords[1]] = 255
num_pepper = np.ceil(prob * image.size * 0.5)
coords = [np.random.randint(0, i-1, int(num_pepper)) for i in image.shape]
sp_image[coords[0], coords[1]] = 0
plt.imshow(sp_image, cmap='gray')
plt.title("Salt & Pepper Noisy Image")
plt.axis('off')
plt.show()
```

```python
plt.hist(sp_image.ravel(), bins=256, range=(0,255), color='green')
plt.title("Histogram - Salt & Pepper Noise")
plt.show()
```

```python
prob = 0.05
impulse_image = np.copy(image)
num_impulse = int(prob * image.size)
coords_x = np.random.randint(0, image.shape[0], num_impulse)
coords_y = np.random.randint(0, image.shape[1], num_impulse)
impulse_values = np.random.choice([0, 255], num_impulse)
impulse_image[coords_x, coords_y] = impulse_values
plt.imshow(impulse_image, cmap='gray')
plt.title("Impulse Noisy Image")
plt.axis('off')
plt.show()
```

```python
plt.hist(impulse_image.ravel(), bins=256, range=(0,255), color='red')
plt.title("Histogram - Impulse Noise")
plt.show()
```
